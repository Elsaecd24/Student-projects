## Plotly

```python
import cupy as cp
import plotly.io as pio
import plotly.graph_objects as go
import random

# Définir le renderer par défaut
pio.renderers.default = 'notebook'

# Fonction pour générer une couleur aléatoire unique
def random_color(existing_colors):
    while True:
        color = f'rgba({random.randint(0, 255)}, {random.randint(0, 255)}, {random.randint(0, 255)}, 1)'
        if color not in existing_colors:
            existing_colors.add(color)
            return color
        
# Nombre de points et de fichiers
num_points = 289
num_files = 1

# Initialiser la figure
fig = go.Figure()

# Charger toutes les données une seule fois
data = [cp.load(f'worm1/track_results/coords_real/coords{j:06}.npy') for j in range(num_files)]

# Convertir les données en tableaux NumPy pour Plotly
data_np = [cp.asnumpy(d) for d in data]

# Initialiser un ensemble pour suivre les couleurs utilisées
used_colors = set()

# Ajouter les trajectoires pour chaque point
for i in range(num_points):
    x = [data_np[j][i, 0] for j in range(num_files)]
    y = [data_np[j][i, 1] for j in range(num_files)]
    z = [data_np[j][i, 2] for j in range(num_files)]
    
    fig.add_trace(go.Scatter3d(
        x=x,
        y=y,
        z=z,
        mode='markers',
        marker=dict(size=4, color=random_color(used_colors)),
    ))

# Configurer les axes et le fond
fig.update_layout(
    scene=dict(
        xaxis=dict(title='X Axis'),
        yaxis=dict(title='Y Axis'),
        zaxis=dict(title='Z Axis'),
        bgcolor='white'
    ),
    width=800,
    margin=dict(r=20, b=10, l=10, t=10)
)

# Afficher la figure avec le renderer 'iframe'
fig.show(renderer="iframe")
```

```python
import cupy as cp
import plotly.io as pio
import plotly.graph_objects as go
from scipy.spatial import Delaunay
import random

# Définir le renderer par défaut
pio.renderers.default = 'notebook'

# Fonction pour générer une couleur aléatoire unique
def random_color(existing_colors):
    while True:
        color = f'rgba({random.randint(0, 255)}, {random.randint(0, 255)}, {random.randint(0, 255)}, 1)'
        if color not in existing_colors:
            existing_colors.add(color)
            return color
        
# Nombre de points et de fichiers
num_points = 289
num_files = 1

# Initialiser la figure
fig = go.Figure()

# Charger toutes les données une seule fois
data = [cp.load(f'worm1/track_results/coords_real/coords{j:06}.npy') for j in range(num_files)]

# Convertir les données en tableaux NumPy pour Plotly
data_np = [cp.asnumpy(d) for d in data]

# Initialiser un ensemble pour suivre les couleurs utilisées
used_colors = set()

# Ajouter les trajectoires pour chaque point
for i in range(num_points):
    x = [data_np[j][i, 0] for j in range(num_files)]
    y = [data_np[j][i, 1] for j in range(num_files)]
    z = [data_np[j][i, 2] for j in range(num_files)]
    
    fig.add_trace(go.Scatter3d(
        x=x,
        y=y,
        z=z,
        mode='markers',
        marker=dict(size=4, color=random_color(used_colors)),
    ))

# Effectuer la triangulation de Delaunay pour générer les triangles
points = cp.concatenate(data, axis=0)
tri = Delaunay(cp.asnumpy(points))

# Ajouter les triangles à la figure
for simp in tri.simplices:
    x = [points[i, 0] for i in simp]
    y = [points[i, 1] for i in simp]
    z = [points[i, 2] for i in simp]
    
    fig.add_trace(go.Mesh3d(
        x=x,
        y=y,
        z=z,
        opacity=0.5,
        color='lightblue'
    ))

# Configurer les axes et le fond
fig.update_layout(
    scene=dict(
        xaxis=dict(title='X Axis'),
        yaxis=dict(title='Y Axis'),
        zaxis=dict(title='Z Axis'),
        bgcolor='white'
    ),
    width=800,
    margin=dict(r=20, b=10, l=10, t=10)
)

# Afficher la figure avec le renderer 'iframe'
fig.show(renderer="iframe")
```

## Mayavi

```python
from mayavi import mlab
import numpy as np

# Créer des données
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 11, 12, 13, 14])
z = np.array([5, 6, 7, 8, 9])

# Créer une figure 3D
mlab.points3d(x, y, z, scale_factor=0.5)

# Afficher la figure
mlab.show()
```
