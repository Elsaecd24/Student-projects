## Bilan et description des forces aérodynamiques
Bilan des forces sur les pales de l'éolienne :
- Poids
- Portance
- Traînée



<p align="center">
  <img src="Schéma Pale éolienne.png">
</p>

**Bilan des forces :**

Force de trainée : $ F_{t}=\frac{1}{2}ρSv_{0}^2C_{x} $ (parallèle à la direction du vent relatif)

Force de portance : $ F_{p}=\frac{1}{2}ρSv_{0}^2C_{y} $ (perpendiculaire à la direction du vent relatif) 

Poids : $P= {m}{g}$ (parallèle à la normale gravitaire)


**Avec:**

Vitesse du vent relatif : $\vec{v_{0}} = v_{0}\vec{e_{x}}$ 

$ ρ $ : masse volumique de l'air 

$ S $ : surface portante de l'aile

$ C_{x} (ou $ $C_{d} ) $ : coefficient de trainée 

$ C_{y} (ou $ $C_{l} ) $ : coefficient de portance

$ g $ : intensité de la pesanteur de valeur

### Hypothèses :
On suppose que la vitesse V est uniforme. Nous nous plaçons dans un écoulement uniforme et stationnaire

De plus nous cherchons à avoir une force de traînée minimale, et une force de portance maximale. Il faut donc que nous choisissions nos coefficients et nos pales en fonction de ces contraintes.

## Remarque :

Parmi les courbes que nous avons présentées précédemment, pour confirmer notre choix il nous faudrait pouvoir nous fier à nos valeurs. Pour cela nous avons besoin de lisser nos courbes avec le code suivant, nous allons donc retracer chacun des rapports et voir à l'aide d'une regression linéaire la courbe la plus intéressante.

```python
def moyenneMobile(tab, N):
    T = np.zeros_like(tab, dtype=np.float64)
    nc = tab.shape[0]
    for i in range(nc):
        if N%2 == 0:
            a, b = i - (N-1)//2, i + (N-1)//2 + 2
        else:
            a, b = i - (N-1)//2, i + (N-1)//2 + 1

        a = max(0, a)
        b = min(nc, b)
        T[i] = np.mean(tab[a:b])
    return T
```

## Calcul du couple délivrable par l’éolienne et estimation de sa puissance nominale.
Pour calculer le couple, nous avons besoin de projeter les deux expressions des forces de portance et de traînée sur l'axe horizontal. Nous allons ensuite intégrer ces forces sur toute la longueur de la pale.

$F_{x}=\cos⍺.dF_{t}(x)+\sin⍺.dF_{p}(x)$

$Avec:$

$dF_{t}=\frac{1}{2}ρv_{0}^2C_{x}Cdl $ 

$dF_{p}=\frac{1}{2}ρv_{0}^2C_{y}Cdl $ 

```python
import numpy as np

C = 6 #longueur de la corde en mètres
Rho = 1.2 #masse volumique de l'air
Cx = 0.0130 #Coefficient de traînée
Cy = 1.0332 #Coefficient de portance
v0 = 6.756 #vitesse du vent en m/s
w = 10*2*np.pi/60 #vitesse de rotation de la pâle en rad/s
L = 47 #longueur de la pale en mètres
alpha = 5*np.pi/180 #angle d'incidence
import numpy as np

print(w)

# Méthode du rectangle

def f(x):
    return (np.cos(alpha)*1/2*Rho*((np.sqrt(v0**2+(x*w)**2))**2*Cx*C*x)+(np.sin(alpha)*1/2*Rho*((np.sqrt(v0**2+(x*w)**2)))**2*Cy*C*x))

def rectangles(a,b,n):
    x=np.linspace(a,b,5)
    S=0
    for i in range(0,n):
        xi=a+(b-a)*i/float(n)
        xj=a+(b-a)*(i+1)/float(n)
        S+=f((xi+xj)/2)*(xj-xi)
    return S

print(rectangles(0,47,1000)*3*10*2*np.pi/60)
print (np.sqrt(v0**2+(40*w)**2))
```

# 2. Modélisation comportementale du mât

## LE MODÈLE DE COMPORTEMENT RETENU POUR LE MÂT


Comme annoncé précédemment, nous avons choisi un mât de hauteur **h=80 mètres**. Le diamètre de la base d’une tour d'éolienne peut-être de plus 5 m en diminuant progressivement pour atteindre environ 3 m au sommet. Nous modélisons donc le mât par un cylindre de diamètre **D=4m**

Le mât est soumis à deux forces verticales qui s'annulent: le poids et la réaction normale. Horizontalement, il est soumis uniquement à la force de trainée de formule $ F_{t}=\frac{1}{2}ρSv_{z}(x)^2C_{x} $ (parallèle à la direction z du vent relatif)

Modélisation du problème: l'impact de la force de trainée sur le mât peut se modéliser par une **charge linéique qui s'applique sur une poutre en flexion** encastrée d'un coté et libre de l'autre. Cette situation se traduit par cette équation:

$EI_{Gz}\frac{d^4z(x)}{dx^4}=Q(x) $

où:

$E$ : le module de Young du matériau

$I_{Gz}$ : le moment quadratique

$z(x)$ : la flèche dirigée selon z et qui varie selon x

$Q(x)$: la charge linéique

Dans notre cas, la charge linéique est la trainée qui varie selon la hauteur à laquelle on se place, c'est-à-dire:

$Q(x)=\frac{F_{t}}{h}=\frac{1}{2}ρ\frac{S}{h}v_{z}(x)^2C_{x}=\frac{1}{2}ρDv_{z}(x)^2C_{x}$ $[\frac{N}{m}] $

$S$: la surface portante du mat = la surface projetée d'un cylindre soit un rectangle de longueur $h$ et de diamètre $D$

Ainsi, déterminer la flèche revient à **établir un profil de vitesse** du vent à différentes hauteurs puis à **intégrer quatre fois la charge linéique**.

## L'IMPACT D'UN BALLOURD SUR LA VOILURE

Dans le cas d’une poutre cylindrique de longueur $L$ :
Les modes propres sont solutions de l’équation suivante :

$cosh(kL)cos(kL)=-1$ avec $k= (\frac{ρπR^2ω^2}{EI})^\frac{1}{4}$

On assimile cette poutre à notre mât, on a donc $L=h=80$ $m$

Pour chaque valeur de kL=x trouvée, on aura :

$ω=\frac{x^2}{L^2R}\sqrt{\frac{EI}{ρπ}}  $ $s ^{-1} $

où $E = E_{acier inoxydable}= 220  GPa $ 

et $ρ = ρ_{acier inoxydable}= 7850 \frac{kg}{m^3} $

On a donc $ω=x^2 \cdot \frac{1}{80^22}\sqrt{\frac{220 \cdot 10^94π}{7850π}}=0,827x^2 $ $s^{-1}$

Afin d'étudier l'impact d'un ballourd sur la voilure, nous allons procéder aux étapes suivantes.

**1) Calcul des modes propres et fréquences propres.**

Trouver les modes propres solutions de l’équation  $cosh(kL)cos(kL)=-1$ revient à résoudre numériquement avec la méthode de Newton l'équation  $cosh(kL)cos(kL)+1=0$

```python
#1 Calcul des fréquences propres. 

# Etape 1: calcul des modes propres 

import numpy as np
import matplotlib.pyplot as plt

def f(z):
    return np.cosh(z)*np.cos(z)+1

def fPrime(z):
    return np.sinh(z)*np.cos(z)-np.cosh(z)*np.sin(z)

kL=np.arange(0,10,1)


#on cherche les intervalles où le signe change
def intervalles():
    L=[]
    for i in range(len(kL)-1):
        if f(kL[i])*f(kL[i+1])<0:
            L=L+[kL[i],kL[i+1]]
    return L
        
print(intervalles())
```

```python
#On cherche les solutions dans les 3 premiers intervalles
def newton(f,fPrime,L):
    x=0
    delta = 1
    while delta > 10**-2:
        for i in range(len(L)): 
            x = -f(L[i])/fPrime(L[i]) + L[i]
            delta = abs(x - L[i])
            L[i] = x
        return x

kL1=np.arange(1,2,0.0001)
kL2=np.arange(4,5,0.0001)
kL3=np.arange(7,8,0.0001)

ModePropre1=newton(f,fPrime,kL1)
ModePropre2=newton(f,fPrime,kL2)
ModePropre3=newton(f,fPrime,kL3)

print(ModePropre1)
print(ModePropre2)
print(ModePropre3)
```

```python
# Etape 2: Calcul des fréquences propres

def FréquencePropre(ModePropre):
    return (ModePropre**2)*0.827

FréquencePropre1=FréquencePropre(ModePropre1)
FréquencePropre2=FréquencePropre(ModePropre2)
FréquencePropre3=FréquencePropre(ModePropre3)

print(FréquencePropre1)
print(FréquencePropre2)
print(FréquencePropre3)
```

**2) Représentation graphique des modes propres.**

```python
#2 Représentation graphique des modes propres.

import matplotlib.pyplot as plt

solutions = [ModePropre1,ModePropre2,ModePropre3]
y = [FréquencePropre1,FréquencePropre2,FréquencePropre3]

plt.title('Représentation graphique des modes propres')
plt.xlabel('kL')
plt.ylabel('ω')

plt.scatter(solutions,y)
plt.show()
```

## Dans le cas d'un vent constant 

* **ENERGIE CINÉTIQUE DU VENT**

Le vent est de l'air en mouvement, se déplaçant d'une zone de hautes pressions vers une zone de basses pressions. Comme tout corps en mouvement, on peut lui associer une energie cinétique.
  
Si on considère que la masse volumique de l'air est constante, on peut dire que l'énergie fournie par le vent est fonction de sa vitesse :


$$E_{c}=\frac{1}{2}.m.v_{vent}^2$$


$ Avec: $

$E_{c}: $ énergie cinétique (en $J$)

$ m:$ masse du volume d'air (en $kg$)

$v_{vent}:$ vitesse instantanée du vent (en $m/s$)


Dans le cas de l’éolien, le volume d'air occupé dépend de la surface balayée par le rotor de l’éolienne. La puissance du vent traversant le rotor correspond à la quantité d’énergie cinétique traversant le rotor chaque seconde.

* ### Calcul de l'Energie cinétique:

On a: $v_{vent}$ = 6,756 $m/s$ et $m$ = 1,204 $kg$ pour une pression de 1 bar et une température de 20°C

Ainsi, $E_{c}$ = $27,48$ $J$

* **PUISSANCE THÉORIQUEMENT RÉCUPÉRABLE**

En considérant un dispositif de récupération de l'énergie cinétique déterminée précedemment, avec une surface $S$ et en faisant l'hypothèse que la vitesse du vent est identique en chaque point de cette surface, le volume d'air qui traverse cette surface en 1s est égale à $v_{vent}S$.

On a: $$ P=E_{c}/t=\frac{1}{2}.D_m.v_{vent}^2=\frac{1}{2}.ρ_{air}.S.v_{vent}^3 $$

$ Avec: $

$P$ : puissance récupérable (en $W$)

$ D_{m}:$ débit massique du volume d'air traversant la surface S en 1s (en $kg/s$)

$S:$  surface du dispositif de récupération (en $m^2$)

$ ρ_{air}:$ masse volumique de l'air (en $kg/m^3$)

### Calcul de la puissance:

Notre mât est tel que: $D$=4 m, $H$=80 m  

Ainsi, $S$=$2\pi\frac{D}{2}H$= 1005 $m^2$

De plus, on sait que $ρ_{air}$= 1,292 $kg/m^3$

Donc, $P$ = $\frac{1}{2}ρ_{air}Sv_{vent}^3 $ = $0,2$ $MW$

Cette valeur est 8 fois plus petite comparée à la puissance de notre éolienne dans la partie 1 (1,6 MW). 

A noter que la masse volumique varie avec différents paramètres :

1. **L'altitude**: plus l'altitude est élevée et plus la masse volumique de l'air sera faible. Une même éolienne produira plus en bord de mer qu'en haute montagne avec la même vitesse de vent.

2. **La température** : plus la température est élevée et plus la masse volumique de l'air sera faible. Une éolienne produira plus lorsque la température ambiante sera faible.

3. **L'humidité** relative de l'air : plus l'air sera humide et plus la masse volumique sera faible mais l'influence de l'humidité relative de l'air sur la masse volumique reste relativement faible.

On décrit normalement les variations du vent sur le site en utilisant une distribution de Weibull comme celle sur l'image ci-dessus. C'est le modèle général qui décrit les variations de la vitesse du vent. Ce modèle permet d'optimiser la conception des éoliennes pour minimiser les coûts liés à la production d'électricité.


On mesure la vitesse moyenne du vent toutes les 10 minutes au moyen d'un anémomètre. Les valeurs obtenues sont alors réparties en différentes vitesses de vent. On peut alors exprimer le potentiel énergétique du site en fonction de la fréquence des différentes classes de vitesse.

On a déterminé précedemment, dans la partie du vent constant, la puissance théoriquement récupérable et puisqu'à présent on est dans le cas d'un vent variable, on peut exprimer graphiquement cette puissance en fonction de la vitesse du vent. 

```python
import math
import numpy as np
import matplotlib.pyplot as plt

Rho1=1.292   #masse volumique de l'air kg/m3
S=829     #surface du pilier de l'éolienne en m2

def P(v):
    return 0.5*Rho1*S*(v**3)

v = np.arange(0, 50, 0.1)

plt.plot(v,P(v))
plt.xlabel('puissance (en W)',color='green')
plt.ylabel('vitesse du vent (en m/s)',color='green')
plt.title("PUISSANCE RÉCUPÉRABLE EN FONCTION DE LA VITESSE DU VENT",color='red')
plt.show()
```
